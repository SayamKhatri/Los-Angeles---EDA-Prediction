Around 13% of Vict Sex field is missing, we predict the missing values as part of our data cleaning

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Filter rows where 'Vict Sex' is not null
data_with_sex = cd_df[cd_df['Vict Sex'].notnull()]

# Selecting a few columns as features
features = ['AREA', 'Vict Age', 'Crm Cd']

X = data_with_sex[features]
y = data_with_sex['Vict Sex']

# Encoding the target variable
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Train a RandomForestClassifier on the data with 'Vict Sex' values
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf.fit(X, y_encoded)

# Calculate accuracy on the entire dataset
y_pred = rf_clf.predict(X)
accuracy = accuracy_score(y_encoded, y_pred)

# Display the accuracy before we fill the missing values with the predicted value
print("Accuracy:", accuracy)

# Predict 'Vict Sex' for rows where it's missing
missing_data_rows = cd_df[cd_df['Vict Sex'].isnull()]
predicted_sex_rf = rf_clf.predict(missing_data_rows[features])

# Inverse transform the predicted values to get the original labels
predicted_sex_labels = le.inverse_transform(predicted_sex_rf)

# Replace missing values in the 'Vict Sex' column with predicted values
cd_df.loc[cd_df['Vict Sex'].isnull(), 'Vict Sex'] = predicted_sex_labels
```

Accuracy of the model is approx 70%

Using a Random Forest model to predict missing 'Vict Sex' values enhances demographic analysis, ensures data completeness, uncovers important factors, reveals correlations, and leverages machine learning for richer insights.

#Q7 Impact of Major Events:

```python
# Define the hypothetical event dates
zero_bail_policy = '2020-04-13' #https://www.thecentersquare.com/california/article_c967f422-621a-11ee-8fbc-8740d492022f.html
covid_lockdown =  '2020-03-19'   # reference link https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7996058/#:~:text=In%20California%2C%20because%20of%20a,announced%20on%20March%2019%2C%202020.
covid_reopen = '2021-06-15' # https://www.dailynews.com/2021/06/14/la-county-joining-california-set-to-largely-reopen-tuesday-june-15-after-more-than-a-year-of-coronavirus-restrictions-heres-what-that-means/


# Group by 'DATE OCC' to get daily crime counts
daily_crime_counts = cd_df.groupby('DATE OCC').size()

# Calculate average crime rates 30 days before and after the Zero Bail Policy
before_zero_bail_policy = daily_crime_counts[pd.Timestamp(zero_bail_policy) - pd.Timedelta(days=30):pd.Timestamp(zero_bail_policy) - pd.Timedelta(days=1)].mean()
after_zero_bail_policy = daily_crime_counts[pd.Timestamp(zero_bail_policy):pd.Timestamp(zero_bail_policy) + pd.Timedelta(days=30)].mean()

# Calculate average crime rates 30 days before and after the COVID Lockdown
before_covid_lockdown = daily_crime_counts[pd.Timestamp(covid_lockdown) - pd.Timedelta(days=30):pd.Timestamp(covid_lockdown) - pd.Timedelta(days=1)].mean()
after_covid_lockdown = daily_crime_counts[pd.Timestamp(covid_lockdown):pd.Timestamp(covid_lockdown) + pd.Timedelta(days=30)].mean()

# Calculate average crime rates 30 days before and after the COVID Reopen
before_covid_reopen = daily_crime_counts[pd.Timestamp(covid_reopen) - pd.Timedelta(days=30):pd.Timestamp(covid_reopen) - pd.Timedelta(days=1)].mean()
after_covid_reopen = daily_crime_counts[pd.Timestamp(covid_reopen):pd.Timestamp(covid_reopen) + pd.Timedelta(days=30)].mean()


before_zero_bail_policy, after_zero_bail_policy, before_covid_lockdown, after_covid_lockdown, before_covid_reopen, after_covid_reopen
```

**Zero Bail Policy (Implemented on April 13, 2020)**

Before: Approximately 489 crimes/day

After: Approximately 538 crimes/day


**COVID Lockdown (Started on March 19, 2020)**

Before: Approximately 571 crimes/day

After: Approximately 492 crimes/day


**COVID Reopen (Started on June 15, 2021)**

Before: Approximately 550 crimes/day

After: Approximately 600 crimes/day

**Based on the events:**

1.   The Zero Bail Policy resulted in a rise in the average daily crime rate.
2.   The COVID Lockdown led to a drop in the average daily crime rate.
3.   The reopening after COVID resulted in a rise in the average daily crime rate.



```python
dates = [zero_bail_policy, covid_lockdown, covid_reopen]
before_rates = [before_zero_bail_policy, before_covid_lockdown, before_covid_reopen]
after_rates = [after_zero_bail_policy, after_covid_lockdown, after_covid_reopen]
events = ['Zero Bail Policy', 'COVID Lockdown', 'COVID Reopen']

# Convert string dates to datetime
dates = [pd.Timestamp(date) for date in dates]

# Create the plot
plt.figure(figsize=(12, 7))
plt.plot(events, before_rates, 'o-', label='Before', color='blue')
plt.plot(events, after_rates, 'o-', label='After', color='red')

# Annotate each event with its date
for i, date in enumerate(dates):
    plt.annotate(date.strftime('%Y-%m-%d'), (events[i], max(before_rates[i], after_rates[i]) + 2),
                 ha='center', fontsize=9, color='gray')

# Labeling and displaying the plot
plt.title('Average Crime Rates Before and After Key Events')
plt.ylabel('Average Crimes/Day')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

The graph showing the average crime rates before and after the key events:

The blue line and points represent the crime rates before each event.

The red line and points represent the crime rates after each event.


**Summary:**
1. **Zero Bail Policy (Implemented on April 13, 2020)**

   **Before**: Approximately 489 crimes/day
   
   **After**: Approximately 538 crimes/day

   **Reasons**:
   - **Release of Detainees**: One of the primary intentions behind a zero bail policy is to reduce the number of people in detention to prevent overcrowding, especially during crises like the COVID-19 pandemic. This can result in the release of individuals who might re-offend.
   - **Reduced Deterrence**: Knowing that they might not be held on bail might embolden some individuals to commit crimes, believing there's a lower risk of long-term detention.
   - **Other Factors**: It's important to consider that other factors not related to the zero bail policy might have influenced crime rates. For instance, economic downturns, unemployment, or other societal factors could contribute to increases in certain types of crimes.

2. **COVID Lockdown (Started on March 19, 2020)**

   **Before**: Approximately 571 crimes/day

An automated method was used to find the best way to predict daily crime counts. The process determined that a specific ARIMA model, denoted as ARIMA(5,1,5)(0,0,0)[0], is the most accurate for this dataset. This took about 120 seconds. This model can help forecast future crime counts effectively.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from statsmodels.tsa.arima.model import ARIMA
# Fit the ARIMA model
model = ARIMA(daily_crime_counts, order=(5,1,5))
model_fit = model.fit()

# Forecast the next 30 days
forecast = model_fit.forecast(steps=30)

# Plot the historical data and the forecast
plt.figure(figsize=(15, 6))
daily_crime_counts.plot(label="Historical Data")
forecast.plot(label="30-day Forecast", color="red")
plt.title('Daily Crime Trend and 30-day Forecast')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Extract forecasted values and confidence intervals
forecast_values = forecast.values
forecast_conf_int = model_fit.get_forecast(steps=30).conf_int()

# dataframe to display forecasted values and confidence intervals
forecast_df = pd.DataFrame({
    'Date': forecast_conf_int.index,
    'Forecasted Crime Counts': forecast_values,
    'Lower Confidence Interval': forecast_conf_int.iloc[:, 0].values,
    'Upper Confidence Interval': forecast_conf_int.iloc[:, 1].values
})

# Display the forecasted values in a table format
forecast_df
```

The table above provides a detailed breakdown of the forecast for the next 30 days:


**Date:** The date of the forecast.

**Forecasted Crime Counts:** The predicted number of crimes for the given date.

**Lower Confidence Interval:** The lower bound of the forecasted value's confidence interval.

**Upper Confidence Interval:** The upper bound of the forecasted value's confidence interval.

**Confidence intervals:**

The plot includes the historical daily crime trend (in blue), the 30-day forecast (in red), and the shaded region representing the confidence intervals (in pink)

```python
# Plot the historical data, forecast, and confidence intervals
plt.figure(figsize=(15, 6))
daily_crime_counts.plot(label="Historical Data")
forecast.plot(label="30-day Forecast", color="red")
plt.fill_between(forecast_conf_int.index,
                 forecast_conf_int.iloc[:, 0],
                 forecast_conf_int.iloc[:, 1],
                 color="pink",
                 alpha=0.3,
                 label="Confidence Interval")
plt.title('Daily Crime Trend, 30-day Forecast, and Confidence Intervals')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

The shaded region provides a visual representation of the range within which we expect the actual number of crimes to fall on each of the forecasted days.
